```python
# Lab 6: Linear Prediction using Linear Regression

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# Generate simple linear data
y = np.array([2, 4, 6, 8, 10, 12, 14, 16])  # Target values
x = np.arange(len(y)).reshape(-1, 1)        # Features: [[0], [1], ..., [7]]
```

```python
print(x,y)
```

```python
# Linear Regression model
model = LinearRegression()
model.fit(x, y)
```

```python
# Predictions
y_pred = model.predict(x)
```

```python
# Evaluation
mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)
```

```python
print("Linear Regression Model")
print(f"Slope (Weight): {model.coef_[0]:.2f}")
print(f"Intercept (Bias): {model.intercept_:.2f}")
print(f"Mean Squared Error: {mse:.2f}")
print(f"R^2 Score: {r2:.2f}")
```

```python
# Plot
plt.scatter(x, y, color='blue', label='Actual')
plt.plot(x, y_pred, color='red', linestyle='--', label='Predicted')
plt.title("Linear Prediction")
plt.xlabel("Input Index")
plt.ylabel("Target Value")
plt.legend()
plt.grid(True)
plt.show()
```
